#### voy a crear otro datada frame merged para la creacion del dash board agregando copias de columnas y cambiando los tipos de datos , todo este proceso va a estar comentado y explicado en el documento ETL con el titulo "PREPARACION DATA FRAME FINAL"

### Creacion de un data frame con columnas aptas para la matriz de correlacion 

#### Elimino columnas inecesarias 

```python
df_merged_only_num = df_merged.drop(['ID', 'LUGAR_DEL_HECHO' , 'FECHA_x' , 
       'Calle', 'Altura', 'Cruce', 'DirecciÃÂ³n Normalizada', 'pos x',
       'pos y', 'VICTIMA_x', 'ROL' , 'FECHA_FALLECIMIENTO'  ] , axis=1)
```

#### Remplazo los valores SD por 0

```python
df_merged_only_num.replace("SD", 0, inplace=True)
```

#### Creacion de dummies para las columnas TIPO_DE_CALLE , ACUSADO Y SEXO 

```python
# Obtén variables ficticias para las columnas categóricas
df_dummies = pd.get_dummies(df_merged_only_num[['TIPO_DE_CALLE', 'ACUSADO', 'SEXO']], drop_first=True , dtype=int)

# Concatena las variables ficticias al DataFrame original
df_masdummies = pd.concat([df_merged_only_num, df_dummies], axis=1)

# Elimina las columnas originales categóricas si lo deseas
df_masdummies.drop(['TIPO_DE_CALLE', 'ACUSADO', 'SEXO'], axis=1, inplace=True)

# Muestra el DataFrame resultante
print(df_masdummies.head())
```

#### Guardo el dataframe con dummies en un archivo .csv por si es necesario su utilizacion en el Dash Board

```python
df_masdummies.to_csv('Dataframes/df_dummies.csv' , index=False)
```

### Creacion de la Matriz de Correlacion 

```python
correlacion_matriz = df_masdummies.corr()
```

```python
print(correlacion_matriz.shape)  # Debería mostrar el número correcto de filas y columnas
print(correlacion_matriz)  # Verifica si todas las correlaciones están presentes
```

```python

plt.figure(figsize=(14, 10))  
sns.heatmap(correlacion_matriz, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

## Conclusion :
1. No se distingui ninguna correlacion muy grande entre las columnas solo hay unas pocas que superan el 0.1 de correlacion y las relaciones a destacar son :

## Graficas

### Obtenemos los tipos de calle con mas accidentes 

```python
Top_tipo_de_calle =df_hechos_clean ['TIPO_DE_CALLE'].value_counts()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(y=Top_tipo_de_calle.index, x=Top_tipo_de_calle.values)
plt.xlabel('Cantidad')
plt.ylabel('TIPO_DE_CALLE')
plt.title('Accidentes por TIPO_DE_CALLE')
plt.xticks(rotation=90)
plt.show()
```

```python
# Assuming Top_tipo_de_calle is your pandas Series with street types as index and their corresponding accident counts as values
labels = Top_tipo_de_calle.index
sizes = Top_tipo_de_calle.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Accidentes por Tipo de Calle')
plt.show()
```

### Observaciones :
La principal observacion es que los hechos ocurren en su mayoria en las avenidas duplicando el valor del segundo lugar donde mas suceden los hechos que es la Calle y otra cosa a destacar es que la gral paz tiene el mismo numero de hechos que la suma de autopista
